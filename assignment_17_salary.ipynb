```python
test.drop(columns='education',inplace=True)
test['Salary'].replace({' >50K':1,' <=50K':0},inplace=True)
test = pd.concat([test,pd.get_dummies(test['workclass'])],axis=1)
test.drop(columns='workclass',inplace=True)
test = pd.concat([test,pd.get_dummies(test['maritalstatus'])],axis=1)
test.drop(columns='maritalstatus',inplace=True)
test = pd.concat([test,pd.get_dummies(test['occupation'])],axis=1)
test.drop(columns='occupation',inplace=True)
test = pd.concat([test,pd.get_dummies(test['relationship'])],axis=1)
test.drop(columns='relationship',inplace=True)
test = pd.concat([test,pd.get_dummies(test['race'])],axis=1)
test.drop(columns='race',inplace=True)
test = pd.concat([test,pd.get_dummies(test['sex'])],axis=1)
test.drop(columns='sex',inplace=True)
test = pd.concat([test,pd.get_dummies(test['native'])],axis=1)
test.drop(columns='native',inplace=True)
cols = list(test.columns)
cols[0],cols[5] = cols[5],cols[0]
test = test[cols]
```

```python
test.info()
```

```python
test.head()
```

```python
X_test = test.iloc[:, 1:].values
y_test = test.iloc[:, 0].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
